This file contains the code for extraction of the data from the structure of the dataset mentioned in the 3rd chapter of thesis.

# 1. Importing libraries used

```python
import pandas as pd
import numpy as np
from xml.dom import minidom
import os, time
```

# 2. Reading files and Extracting the data

```python
#Saving the location of the dataset
root_path = "E:\\MS DA\\Major Project Nuig\\Archive\\"

#Initialising the path of the file
dirList = os.listdir(root_path)

ball_data = [] #to store ball data
frame_game_data = [] #to store frames data

for d in dirList:
    if (os.path.isdir(root_path + str(d)) == True and d != "__MACOSX" and (d == "Round 1 Tracking Data")): #it has to be changed to extract all the rounds data
        print("Stage 1: "+ root_path + str(d))
        
        #Iterating over the Round folder
        for r in os.listdir(root_path + str(d)):
            if os.path.isdir(root_path + str(d) + "\\"+ str(r)) == True:
                #Iterating over game folders
                for f in os.listdir(root_path + str(d) + "\\" + str(r)):
                    file_name_path = root_path + str(d) + "\\" + str(r) + "\\" + f
                    if(f.split(".")[-1] == "xml"): #checking file extension
                        #Storing XML data into variables
                        mydoc = minidom.parse(file_name_path)
                        xml_match_attr = mydoc.getElementsByTagName('match')[0]                        
                        tracking_area_y = xml_match_attr.attributes['fTrackingAreaYSizeMeters'].value
                        tracking_area_x = xml_match_attr.attributes['fTrackingAreaXSizeMeters'].value
                        pitch_size_y = xml_match_attr.attributes['fPitchYSizeMeters'].value
                        pitch_size_x = xml_match_attr.attributes['fPitchXSizeMeters'].value
                        match_date = xml_match_attr.attributes['dtDate'].value.split(" ")[0]
                        match_time = xml_match_attr.attributes['dtDate'].value.split(" ")[1]                        
                        game_id = xml_match_attr.attributes['iId'].value
                        first_period_frame_start = xml_match_attr.childNodes[0].attributes['iStartFrame'].value
                        first_period_frame_end = xml_match_attr.childNodes[0].attributes['iEndFrame'].value
                        second_period_frame_start = xml_match_attr.childNodes[1].attributes['iStartFrame'].value
                        second_period_frame_end = xml_match_attr.childNodes[1].attributes['iEndFrame'].value

                    elif(f.split(".")[-1] == "dat"):
                        for lines in open(file_name_path, 'r'):
                            item = lines.rstrip()
                            item_split = item.split(":")
                            ball_data.append((str(r).split("g")[-1] + "," + item_split[0] + "," + item_split[2]).split(","))
                    elif(f.split(".")[-1] == "csv" and "Splits" in f):
                        for csv_lines in open(file_name_path, 'r'):
                            csv_item = csv_lines.rstrip()
                            if " vs " in csv_item:
                                teams = csv_item.split(" vs ")
                                break;
                frame_game_data.append([match_date,match_time,game_id,tracking_area_y,tracking_area_x,pitch_size_y,pitch_size_x,first_period_frame_start,first_period_frame_end,second_period_frame_start,second_period_frame_end, teams[0], teams[1]])                             
```

## Combining Frames Data With Ball Data

```python
print(len(ball_data), len(frame_game_data))
```

```python
#Merging Ball possesion with frame data
updated_data = []
for i in range(len(frame_game_data)):
    print("Processing Game: ", frame_game_data[i][2])
    for bd in ball_data:
        if(frame_game_data[i][2] == bd[0]):
            updated_data.append(frame_game_data[i] + bd)
```

# Storing The Data In CSV File

```python
df_pud = pd.DataFrame(updated_data, columns=["match_date","match_time","game_id","tracking_area_y","tracking_area_x","pitch_size_y","pitch_size_x","first_period_frame_start","first_period_frame_end","second_period_frame_start","second_period_frame_end", "teama", "teamb", "game_id1", "frame_no", "ball_x", "ball_y", "ball_p", "ball_z", "ball_holding_team", "ball_status", "remove"])
```

```python
#Droping unnecessary column
df_pud = df_pud.drop(columns = ["remove"])
```
